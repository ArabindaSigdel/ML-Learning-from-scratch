# Pandas

**Data Preprocessing** in **Machine Learning** using Pandas

**Data preprocessing** is required tasks for cleaning the data and making it suitable for a machine learning model which also increases the accuracy and efficiency of a machine learning model

**Pandas** is an open source Python package that is most widely used for data science/data analysis and machine learning tasks. It is built on top of another package named Numpy, which provides support for multi-dimensional arrays. Preprocessing is the process of doing a pre-analysis of data, in order to transform them into a standard and notmalized format

Preprocessing involves the following aspects
- Missing values
- Data standardization
- Data normalization
- Data minning

```python
# identifying null values variable
df.isnull()
```

```python
isnull_count = df.isnull().sum()
print("Total number of null variable: \n", isnull_count)
```

```python
df
```

```python
# replace single quote with blank and make the date formatted
df["Date"] = df["Date"].replace("'", "", regex = True)
df
```

```python
# Handle null values, return dataframe with no empty cell
# we can use df.dropna(). (removes empty cells but will not change the data frame)
# if we want dataframe to be changed without creating new variable, simply use inplace = True

newdf = df.dropna()
newdf
```

```python
newdf.isnull().count()
```

```python
# another way of handling null values is by replacing the null values
df.fillna(130, inplace = True)
df
```

```python
df = pd.read_csv("/content/pandas_dataset.csv")
df["Calories"].fillna(130, inplace = True)
df
```

```python
#fill specific with mean median and mode
import pandas as pd
df = pd.read_csv("/content/pandas_dataset.csv")
calories_mean = df["Calories"].mean()
df["Calories"].fillna(calories_mean, inplace=True)
df
```
